# 3-1: The Graphs of Odds, Logit and Sigmoid

## Code.3-1-1: The Graphs of Odds and Logit

```python
# numpy로 그래프 그리기
import numpy as np
import tensorflow as tf

import matplotlib.pyplot as plt
plt.style.use('seaborn')

p_np = np.linspace(0.01, 0.99, 100)
p_tf = tf.linspace(0.01, 0.99 ,100)

odds_np = p_np/(1-p_np)
odds_tf = p_tf/(1-p_tf)

logit_np = np.log(odds_np)
logit_tf = tf.math.log(odds_tf)

fig, axes = plt.subplots(2, 1, figsize=(15, 10), sharex=True)

# 그래프그리기
axes[0].plot(p_np, odds_np)
axes[1].plot(p_np, logit_np)

# odds
xticks = np.arange(0, 1.1, 0.1)
axes[0].tick_params(labelsize=15)
axes[0].set_xticks(xticks)
axes[0].set_ylabel('Odds', fontsize=20, color='darkblue')

# logit
axes[1].tick_params(labelsize=15)
axes[1].set_xticks(xticks)
axes[1].set_ylabel('Logit', fontsize=20, color='darkblue')
# 두 그래프의 xlabel의 뜻
axes[1].set_xlabel('Probability', fontsize=20, color='darkblue')
```

```python
# tensorflow로 그래프 그리기
import numpy as np
import tensorflow as tf

import matplotlib.pyplot as plt
plt.style.use('seaborn')

p_np = np.linspace(0.01, 0.99, 100)
p_tf = tf.linspace(0.01, 0.99 ,100)

odds_np = p_np/(1-p_np)
odds_tf = p_tf/(1-p_tf)

logit_np = np.log(odds_np)
logit_tf = tf.math.log(odds_tf)

fig, axes = plt.subplots(2, 1, figsize=(15, 10), sharex=True)

# 그래프그리기
axes[0].plot(p_tf, odds_tf.numpy())
axes[1].plot(p_tf, logit_tf.numpy())

# odds
xticks = np.arange(0, 1.1, 0.1)
axes[0].tick_params(labelsize=15)
axes[0].set_xticks(xticks)
axes[0].set_ylabel('Odds', fontsize=20, color='darkblue')

# logit
axes[1].tick_params(labelsize=15)
axes[1].set_xticks(xticks)
axes[1].set_ylabel('Logit', fontsize=20, color='darkblue')
# 두 그래프의 xlabel의 뜻
axes[1].set_xlabel('Probability', fontsize=20, color='darkblue')
```

## Code.3-1-2: The Graphs of Sigmoid

```python
import tensorflow as tf

from tensorflow.keras.layers import Activation

X = tf.linspace(-30, 30, 100)
sigmoid = Activation('sigmoid')(X)  # 만드는 동시에 통과까지

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(X.numpy(), sigmoid.numpy()) # sigmoid는 잘 훈련하면 거의 0 or 1에 근사값만 나옴
```

# 3-2: Logistic Regression Models

## Code.3-2-1: Single-variate Logistic Regression Models

```python
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.layers import Dense

plt.style.use('seaborn')

# data sample i개, feature z개
X = tf.random.normal(shape=(100, 1))
dense = Dense(units=1, activation='sigmoid')

Y = dense(X)
print(Y.shape)

fig, ax = plt.subplots(figsize=(7, 7))
ax.scatter(X.numpy(), Y.numpy())
```

## Code.3-3-2: Multi-variate Logistic Regression Models (여러개의 feature를 받는 모델)

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense

plt.style.use('seaborn')

# data sample i개, feature z개
X = tf.random.normal(shape=(100, 5))
dense = Dense(units=1, activation='sigmoid')

Y = dense(X)
print(Y.shape)  # Dense Layer에 unit(neuron)이 하나여서 100 x 1 
```

# 3-3: Binary Classifier with Dense Layers

## Code.3-3-1: Binary Classifier with Dense Layers

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(units=10, activation='relu'))
model.add(Dense(units=5, activation='relu'))
model.add(Dense(units=1, activation='sigmoid')) # 얘가 binary classifier를 만들어줌 (마지막에 1개의 뉴런, sigmoid를 통과해 확률(p)만 결과값)
```
